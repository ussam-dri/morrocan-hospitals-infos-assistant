````python
import os
import pandas as pd

folder_path = r"D:\PorjectTech\CHATHOPOA\backend\Llms_replissage_bd\shema_1\csv"

dfs = []

for file_name in os.listdir(folder_path):
    file_path = os.path.join(folder_path, file_name)
    
    # Vérifier que c'est un fichier et qu'il finit par .csv
    if os.path.isfile(file_path) and file_name.lower().endswith(".csv"):
        print("Lecture :", file_path)
        df = pd.read_csv(file_path, encoding='utf-8')
        dfs.append(df)
    else:
        print("Ignoré :", file_path)

# Combiner tous les CSV en un seul DataFrame
if dfs:
    df_all = pd.concat(dfs, ignore_index=True)
    print("✔ Tableau final chargé ")
else:
    print("Aucun fichier CSV valide trouvé")
````

````python
import pandas as pd
from langchain.docstore.document import Document
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.embeddings import HuggingFaceEmbeddings
from tqdm import tqdm

# ================= 2. Combiner toutes les colonnes =================
texts = df_all.fillna("").apply(lambda row: " ".join(row.astype(str)), axis=1).tolist()

# ================= 3. Définir le Text Splitter =================
splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,        # taille de chaque chunk
    chunk_overlap=50       # chevauchement entre chunks
)

# ================= 4. Créer les documents découpés =================
docs = []
for idx, t in enumerate(tqdm(texts, desc="Splitting text")):
    for chunk in splitter.split_text(t):
        docs.append(Document(page_content=chunk, metadata={"row_id": idx}))

print(f"Nombre total de chunks : {len(docs)}")

# ================= 5. Embeddings =================
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# ================= 6. Créer le vector store Chromadb =================
vector_store = Chroma.from_documents(
    documents=docs,
    embedding=embeddings,
    collection_name="hopitaux_collection",
    persist_directory="./chromadb_data"
)

# ================= 7. Sauvegarder le vector store =================
vector_store.persist()
print("✅ Tous les chunks ont été stockés dans Chromadb !")


````

````python
import google.generativeai as genai
from langchain.vectorstores import Chroma
from langchain_huggingface import HuggingFaceEmbeddings

# 1. Configuration de Gemini
genai.configure(api_key="YOUR_API_KEY") 

model = genai.GenerativeModel("models/gemini-2.5-pro")

# 2. Recharger le Vector Store Chroma
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vector_store = Chroma(
    collection_name="hopitaux_collection",
    persist_directory="./chromadb_data",
    embedding_function=embeddings
)

# 3. Question de l'utilisateur
query = "donner en final toutes les champs en csv"

# 4. Récupérer les passages pertinents du Vector Store
docs = vector_store.similarity_search(query, k=200)  # k = nombre de chunks à récupérer

# 5. Construire le prompt RAG
context = "\n\n---\n\n".join([d.page_content for d in docs])

prompt = f"""
Tu es un expert en ingénierie de données et en structuration de bases de données. 
Tu disposes du contexte suivant provenant de notre base vectorielle ChromaDB :

{context}

Ta tâche est la suivante :
1. Éliminer complètement les champs inutiles : `web_scraper`, `links`, `pagination`.
4. Structurer le résultat sous forme json :
   - `hopital_name`
   - `categorie`
   - `adresse`
   - `telephone`
   - `code_isic`
   - `specialite`
5. Retourner le résultat sous **format JSON**, prêt à être importé dans une base de données.

Important : ne pas inventer d’informations, ne corriger que ce qui est clair à partir du contexte fourni.
"""

# 6. Génération par Gemini
response = model.generate_content(prompt)

print("\n Réponse Gemini :\n")
print(response.text)
````

````python
import re
import json

# Extraire le JSON encadré par ```json ... ```
match = re.search(r"```json(.*?)```", response.text, re.DOTALL)
if match:
    json_str = match.group(1).strip()  # Le texte JSON pur
    data = json.loads(json_str)        # Convertir en dictionnaire Python
    print(json.dumps(data, indent=2, ensure_ascii=False))
else:
    print("Aucun JSON trouvé dans la réponse")
````

````python
import re
import json
import pandas as pd

# Extraire le JSON encadré par ```json ... ```
match = re.search(r"```json(.*?)```", response.text, re.DOTALL)
if match:
    json_str = match.group(1).strip()
    data = json.loads(json_str)  # dictionnaire ou liste Python

    # Sauvegarder en fichier JSON
    with open("hopitaux.json", "w", encoding="utf-8") as f:
        json.dump(data, f, indent=2, ensure_ascii=False)
    print("JSON sauvegardé dans hopitaux.json")

    # Sauvegarder en CSV si c'est une liste de dict
    if isinstance(data, list):
        df = pd.DataFrame(data)
        df.to_csv("hopitaux.csv", index=False, encoding="utf-8-sig")
        print(" CSV sauvegardé dans hopitaux.csv")
else:
    print("Aucun JSON trouvé dans la réponse")
````
